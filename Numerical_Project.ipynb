```python
#2.1: Euler Method
#We want to compute approximate solutions for evenly spaced points x0, x1, x2, ...xn.
#Plot each solution and compare your results both graphically and numerically.

#solve the following initial value problems:
#1. y′ = 2x^2 + 3y^2 − 2; y(2) = 1; h = 0.05
#2. y′ = y + √︁ x2 + y2; y(0) = 1; h = 0.1
#3. y′ = (1 + x)/(1 − y2); y(2) = 3; h = 0.1

import numpy as np

#given functions and h values
#fx1 = 2*(x**2) + 3*(y**2) - 2
#fx2 = y + np.sqrt((x**2)+(y**2))
#fx3 = (1 + x)/(1 − y**2)
h1 = 0.05
#h2 = 0.1
#h3 = 0.1

#y1 is y(2)=1
#y2 is y(0)=1
#y3 is y(2)=3

#initial y and x values
x_0 = 2
y_0 = 1

#x and y arrays
x = []
y = []

#defined Euler Method running thru iterations
def euler_method(iterations):
    y_n = y_0
    y.append(y_n)
    for i in range(iterations):
        x_n = x_0 + i*(h1)
        x.append(x_n)
        #Euler's: y_i+1 = y(x_i) + hf(x_i,y(x_i))
        y_x = y_n + h1*((2*(x_n)**2) + (3*(y_n)**2) - 2)
        y.append(y_x)
        y_n = y_x
        i+=1
    y.pop()
    print(x,y)
```

```python
#printing x and y arrays for Euler's Method
print(euler_method(15))
```
